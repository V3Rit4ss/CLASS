```python
import tensorflow.keras.utils as utils #분류분석시 라벨링.
from tensorflow.keras.models import Sequential # 모델 생성시
from tensorflow.keras.layers import Dense, Activation # 입,출력값 과 Activation이용하려고
import numpy as np # csv를 np로 읽어들일려고.
```

# 1.데이터 셋 생성

```python
 # np.loadtxt('') 넘파이배열로 가져오는. csv 데이터안에 #은 주석처리 되어서 생략되어진다.
dataset = np.loadtxt('data/pima-indians-diabetes.csv', delimiter=',')
dataset
```

```python
dataset.shape # 770 에서 #인 주석9개를 뺴면 768.
```

```python
# 학습용 데이터(훈련데이터)
x_train = dataset[:700, :8] # <-700개 행의 모든것 0~8열 앞까지.
y_train = dataset[:700, 8]

#테스트 데이터.
x_test = dataset[700:, :8] 
y_test = dataset[700:, 8]
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# 라벨링 0 -> [1,0] / 1 -> [0,1]. 하지만 여기서는 안할것이다.
```

# 2. 모델 구성하기

```python
#학습을 시킬때 epochs를 늘려도 정확도가 안늘어나면 Dense 값을 수정해야한다.
# 그래도 안늘어나면 분류분석으로 해봐야한다.
model = Sequential()
model.add(Dense(12, input_dim=8 , activation="relu")) # input_dim = 8 ,최종 출력은 1
model.add(Dense(8, activation="relu")) # 처음쓴 input_dim 값을 써야하는데 생략가능.
model.add(Dense(1, activation="sigmoid"))
```

# 3. 학습과정 설정.

```python
# 분류분석x.
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

# 4. 학습시키기

```python
hist = model.fit(x_train, y_train, epochs=1500, batch_size=64)
```

```python
hist.history["loss"]
```

# 5. 모델 평가하기.

```python
score = model.evaluate(x_test, y_test)
```

```python
print("loss : %s\n accuracy : %.2f" % (score[0], score[1]))
```
