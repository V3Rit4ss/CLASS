**2) subplots() 함수로 서브플롯 그리기**

```python
x = np.arange(0, 10, 0.01)
x.shape
```

```python
# 방법1 (2행2열 서브플롯)
plt.subplot(2,2, 1); plt.plot(x, np.sin(x))  # ; 한줄에 여러문법 실행가능
plt.subplot(2,2, 2); plt.plot(x, np.cos(x)) 
plt.subplot(2,2, 3); plt.plot(x, np.tanh(x)) 
plt.subplot(2,2, 4); plt.plot(x, np.sin(x)*np.cos(x)) 
plt.show()
```

```python
# 방법2 (2행 2열 서브플롯's)
# fig, axes = plt.subplots(nrows= 2, ncols= 2) # 2행2열 짜리 빈그래프만 생성.
fig, axes = plt.subplots(nrows= 2, ncols= 2, figsize=(10,6)) # 그래프크기를 조정.
# axes # 2행 2열 의 axes로 나온다.
axes[0,0].plot(x, np.sin(x), color='r', linewidth=10) 
#color='r' : 색상 , linewidth=10 : 라인의 굵기.
axes[0,0].plot([0,1,2,3],[-1,-0.5,0,0.5], 'k:') # 윗 문법의 그래프와 같은곳에 그려진다.
axes[0,1].plot(x, np.cos(x), 'c--') # c -> 청록색, -- : (대시드)
axes[1,0].plot(x, np.tanh(x), 'g')
axes[1,1].plot(x, np.sin(x)*np.cos(x), 'r-.')
plt.show()
```

```python
#방법3 (2행 2열 서브플롯)
x = np.arange(0,7, 0.01)
x.shape
```

```python
# def sin_cos(x):
#     return np.sin(x)*np.cos(x)
# func_list = [np.sin, np.cos, np.tanh, sin_cos]
func_list = [np.sin, np.cos, np.tanh, lambda x : np.sin(x)*np.cos(x)]
fmt_list = ['k:', 'c--', 'g', 'r--']
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
# axes# 2행 2열. => 2차원.
# list(axes.flat) # 2차원을 1차원으로 한 이유, 중첩for문을 안쓰기 위해서.
for i, ax in enumerate(axes.flat):
    ax.plot(x, func_list[i](x), fmt_list[i])
```

```python
fig, axes = plt.subplots(ncols=4, figsize = (18, 3)) #ncols=4,1행 4열. nrows =1(기본값) 
for i, ax in enumerate(axes): # 어차피 1차원이라 flat 안함.
    ax.plot(x, func_list[i](x), fmt_list[i])
```

```python
fig, axes = plt.subplots(nrows=4, figsize = (10,10)) #4행 1열. ncols=1 (기본값)
for i, ax in enumerate(axes):
    ax.plot(x, func_list[i](x), fmt_list[i])
```

## 2.5 다양한 그래프 그리기 (ex)

**2) bar(), axvline(), axhline(), ... **<br>

```python
x = np.linspace(0, 10)
y = np.cos(x)
fig, axes = plt.subplots(2,2, figsize=(10, 5))
axes[0,0].scatter(x,y, s=5, c='r', marker='o')
axes[0,1].fill(x,y, c='b') # 채우기.
axes[1,1].fill_between(x,y, color='r') #fill_between의 색상은 color.
axes[1,1].plot([-1,5],[-1,1], 'ro-')
```

```python
fig, axes = plt.subplots(2,2)
axes[0,0].bar([1,2,3],[3,4,5], color='r')
axes[0,0].plot([1,2,3],[3,4,5], 'k')
axes[0,1].bar([0.5, 1, 2.5], [0,1,2])
axes[0,1].plot([0.5, 1, 2.5], [0,1,2], 'g') # 0이라서 안보이는 bar가 있음.
axes[1,0].axvline(0.5)
axes[1,1].axhline(0.75)
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
#빈도표
axes[0,0].hist(iris_df["sepal_length"],bins=20, color='r')
#누적빈도표(cumulative=True)
axes[0,1].hist(iris_df["sepal_length"],bins=20, cumulative=True)
#가로 표현 빈도표
axes[1,0].hist(iris_df["sepal_length"],bins=10, orientation='horizontal')
# 내부가 비어있는 빈도표
axes[1,1].hist(iris_df["sepal_length"],bins=10, histtype='step')
plt.show()
```

**4) 축과 눈금**

```python
import numpy as np
x = np.linspace(0,10 ,100)
y = np.cos(x)
fig, axes = plt.subplots(2, 2 , figsize=(10,5))
# plt.subplots_adjust() 을 사용하여 서브플롯간의 간격 변경
#간격을 조정하면 figsize의 한계점까지 줄고 늘고 하는데 그래프의 크기가 줄어줄수도 있다.
plt.subplots_adjust(hspace=0.4, wspace=0.3) # hspace 는 위아래 여백, wspace는 옆여백.
plt.suptitle("Main title") # 전체 타이틀.
axes[0,0].scatter(x, y, marker='.') # title="An Example Axes" 서브타이틀
axes[0,0].set(title="An Example Axes", ylabel="y-axis", xlabel="x축",
              xlim=[0,12], # x축의 공간을 변경
              ylim=[-2,2]) # y축의 공간을 변경
axes[0,1].scatter(x, y, marker='*', c='b')
axes[0,1].set_xlim(2,6)
axes[0,1].set_xlabel('x축 0~5')
axes[0,1].set_ylim(-1.5, 1.5)
axes[0,1].set_ylabel('y축 -1.5 ~ 1.5')
axes[1,0].scatter(x, y, marker='^', c='c')
# x축의 눈금위치 변경
# set_xticks(range(11))0부터 10까지 -> range , list 가능.
axes[1,0].set_xticks(range(0,11, 3)) # range(0,11, 2) 0~10까지 2씩 증가해라.
#눈금의 내용 변경. ([])안의 갯수와 x축눈금의갯수와 동일해야한다. 
axes[1,0].set_xticklabels([-100, 3, '6개','foo'])
axes[1,0].set_yticks([-2,0,1,2]) # y축의 눈금중 특정만 보이게.
axes[1,0].set_yticklabels(['-2개','제로',1,'MAX'])
# axes[1,0].set(xticks=range(0,11,3)) 가능.
axes[1,1].scatter(x, y, c='r') # marker 지정안하면 기본값으로 출력.
axes[1,1].set(xticks=[0,2.5,5,7.5,10], xticklabels=['Min',2.5,5,7.5,'MAX'],
              title='test')
axes[1,1].grid(True) # 그래프 백그라운드에 격자 그리드 표현
axes[1,1].spines["top"].set_visible(False)
axes[1,1].spines["bottom"].set_position(("outward",10))
plt.show()
# https://matplotlib.org/3.1.0/api/axis_api.html 참조.
```

**5) 축 공유**

```python
import numpy as np
x = np.arange(0, 10) #0부터 9까지
y1 = 0.5*x**2 # 0.5*(x**2) 제곱이 우선순위가 높다.
y2 = -1 * y1
```

```python
fig, ax1 = plt.subplots() # 기본값이 1행1열 그래프 객체 생성.
ax1.plot(x, y1, 'g^:')
ax1.set(xlabel='X data', ylabel="y1 data")
#그래프 범위가 벗어나지 않게. 새로운 그래프를 그리고 축을 공유하려고 한다.
ax1.plot([0,1,2,3,4],[0,-10,-20,-30,-40]) 
# [0,-10,-20,-30,-40] 범위가 완전히 다르면 그래프의 의미가 사라진다.
```

```python
fig, ax1 = plt.subplots() # 기본값이 1행1열 그래프 객체 생성.
#전체 타이틀.
#plt.suptitle("Main title")
ax1.plot(x, y1, 'g^:')
ax1.set(xlabel='X data', ylabel="y1 data", title='축 공유')
#그래프 범위가 벗어나지 않게. 새로운 그래프를 그리고 축을 공유하려고 한다.
ax2 = ax1.twinx() # 축을 공유하는 ax2
ax2.plot(x, y2, 'b--')
ax2.set_ylabel('y2 data', color='b')
ax3 = ax1.twiny() # ax1과 x축을 공유하는 ax3
ax3.plot(-x, y1, 'ro-')
ax3.set_xlabel('-x, data',color='r')
plt.show()
```
